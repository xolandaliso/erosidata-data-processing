```python
fig = pl.gcf()
nrows = 3
ncols = 3
fig.set_size_inches(ncols * 6, nrows * 4)

def func(x, a, x0, sigma): 
    return a*np.exp(-(x - x0)**2/(2*sigma**2)) 
  
      
for i in range(len(curves_6_9keV)):
    # Set up subplot; subplot indices start at 1
    
    f_eband_lc = fits.open(w_dir[0]+'/light_curves/'+curves_6_9keV[i])
    sp = pl.subplot(nrows, ncols, i + 1)
    hist = pl.hist(f_eband_lc[1].data['RATE'], bins=15, color='k', histtype='step')

    x, y = (hist[1][1:] + hist[1][:-1])/2, hist[0]
    
    a, x0, sig = np.max(y), np.mean(x), np.std(x)
    
    popt, cov = curve_fit(func, x, y, p0=[a, x0, sig])
    x_mock = np.linspace(0.2, np.max(x) + .1, 1000)
    pl.plot(x_mock, func(x_mock, *popt), 'r-')
    pl.axvline(popt[1] + 3*popt[-1], color = 'r', ls = '--')
    pl.axvline(popt[1] - 3*popt[-1], color = 'r', ls = '--')
    pl.text(.2, np.max(func(x_mock, *popt)), r'$\rm 3\sigma =$'+str(np.round(3*popt[-1],2)))
    if i > 2: pl.xlabel(r'$\rm rate\ (cts/s/deg^2)$', fontsize = 14)
    if i == 0 or i == 3: pl.ylabel(r'$\rm counts$', fontsize = 14)
```

#### $\texttt{spectra for each of the TMs using the srctool}$

The spectra is created around the bright region of the cluster using a circle and an annulus that is offset as source and background regions, respectively

```python
fig = pl.gcf()
nrows = 3
ncols = 3
fig.set_size_inches(6, 6)

source_rmf = [ x for x 
                in sorted(os.listdir('new_spectra/')) if 'RMF' in x ]

source_spec = [ x for x 
                    in sorted(os.listdir('new_spectra/')) if 'SourceSpec' in x ]

bk_spec = [ x for x 
                    in sorted(os.listdir('new_spectra/')) if 'BackgrSpec' in x ]

for i in range(len(source_rmf)):

    source_spectrum = fits.open(w_dir[0]+'/new_spectra/'+source_spec[i])['SPECTRUM'].data

    bk_spectrum = fits.open(w_dir[0]+'/new_spectra/'+bk_spec[i])['SPECTRUM'].data

    rmf_data = fits.open(w_dir[0]+'/new_spectra/'+source_rmf[i])['EBOUNDS'].data

    # Calculate the energy for each channel (midpoint of the bin)
    energies = (rmf_data['E_MIN'] + rmf_data['E_MAX']) / 2

    # Define the energy range of interest
    energy_min = 0.2
    energy_max = 9.0

    mask = (( energies >= 0.2 ) & \
                                             ( energies <= 9.0 ) )
    filtered_counts_obs = source_spectrum['COUNTS'][mask]
    filtered_energies_obs = energies[mask]
    filtered_bk = bk_spectrum['COUNTS'][mask]

    sp = pl.subplot(nrows, ncols, i + 1)
    pl.plot(filtered_energies_obs, filtered_counts_obs, 'r-')
    pl.plot(filtered_energies_obs, filtered_bk, 'k-')
    if i > 2: pl.xlabel(r'$\rm E\ (keV)$', fontsize = 14)
    if i == 0 or i == 3: pl.ylabel(r'$\rm counts$', fontsize = 14)
pl.subplots_adjust(right=1.9, top = 1.5, wspace=0.3, hspace=0.3)
```
