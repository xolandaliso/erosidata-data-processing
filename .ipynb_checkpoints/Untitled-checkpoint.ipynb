```python
import os
import pylab as pl 
import numpy as np
from astropy.io import fits
from scipy.optimize import curve_fit

w_dir = !pwd

light_curves_f = [ x for x 
                          in sorted(os.listdir('light_curves/')) if 'full' in x ]

curves_6_10keV = [ x for x 
                            in sorted(os.listdir('light_curves/')) if not 'full' in x ]

fig = pl.gcf()
nrows = 3
ncols = 3
fig.set_size_inches(ncols * 6, nrows * 4)

for i in range(len(light_curves_f)):
    # Set up subplot; subplot indices start at 1
    f_eband_lc = fits.open(w_dir[0]+'/light_curves/'+light_curves_f[i])
    
    sp = pl.subplot(nrows, ncols, i + 1)
    time = np.linspace(0, 50, len(f_eband_lc[1].data['RATE']))
    pl.plot(time, f_eband_lc[1].data['RATE'], 'k-')
    if i > 2 : pl.xlabel('time (ks)', fontsize = 14)
    if i == 0 or i == 3 : pl.ylabel(r'$\rm rate\ (cts/s/deg^2)$', fontsize = 14)
    
```

```python
fig = pl.gcf()
nrows = 3
ncols = 3
fig.set_size_inches(ncols * 6, nrows * 4)

def func(x, a, x0, sigma): 
    return a*np.exp(-(x - x0)**2/(2*sigma**2)) 
  
      
for i in range(len(light_curves_f)):
    # Set up subplot; subplot indices start at 1
    
    f_eband_lc = fits.open(w_dir[0]+'/light_curves/'+curves_6_10keV[i])
    sp = pl.subplot(nrows, ncols, i + 1)
    hist = pl.hist(f_eband_lc[1].data['RATE'], bins=15, color='k', histtype='step')

    x, y = (hist[1][1:] + hist[1][:-1])/2, hist[0]
    
    a, x0, sig = np.max(y), np.mean(x), np.std(x)
    
    popt, cov = curve_fit(func, x, y, p0=[a, x0, sig])
    x_mock = np.linspace(0.2, np.max(x) + .1, 1000)
    pl.plot(x_mock, func(x_mock, *popt), 'r-')
    pl.axvline(popt[1] + 3*popt[-1], color = 'r', ls = '--')
    pl.axvline(popt[1] - 3*popt[-1], color = 'r', ls = '--')
    pl.text(.2, np.max(func(x_mock, *popt)), r'$\rm 3\sigma =$'+str(np.round(3*popt[-1],2)))
    if i > 2: pl.xlabel(r'$\rm rate\ (cts/s/deg^2)$', fontsize = 14)
    if i == 0 or i == 3: pl.ylabel(r'$\rm counts$', fontsize = 14)
```
